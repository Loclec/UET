### B3

```python
a=input("Nhap mot ky tu: ")
if a.islower():
    b= a.upper()
    print(f"Chu cai in hoa la: {b} ")
else:
    b= a.lower()
    print(f"Chu cai in thuong la: {b} ")
```

### B4

```python
c=input("Nhap mot ky tu: ")
if c.isalpha():
    print(f"{c} la ky tu alphabet")
else:
    print(f"{c} khong phai la ky tu alphabet")
```

### B5

```python
c= input("Nhap mot chu cai in hoa: ")
if c.isupper():
    d= ord(c)-1
    if d < ord('A'):
        d= ord('Z')
    print(f"Ky tu truoc do la: {chr(d).lower()}")
else: 
    print("Vui long nhap chu cai in hoa")
    

        
```

### B6

```python
abc= map(int, input("Nhap do dai 3 canh tam giac(cach boi dau cach): ").split())
a, b, c= abc
if a+b>c and a+c>b and b+c>a:
    print("3 canh tren tao thanh 1 tam giac")
else:
    print("3 canh tren khong tao thanh 1 tam giac")
```
